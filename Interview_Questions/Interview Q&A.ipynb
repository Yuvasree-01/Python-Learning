```python
import pandas as pd
import numpy as np
# Students = { "Name": ["Yuvasree","Hari","Kamal","Munu"],
#              "Age":[ 23,24,np.nan,22],
#              "Sex": ["Female","Male","Male","Male"],
#              "Qualification":["B.tech","B.E","B.Tech","B.E"],
#              "Experienced": ["8 Months","1 year","2 years",np.nan]}
# pd.DataFrame(Students)

data = {
    'Name': ['Aishwarya', 'Bikram', 'Cheetah', 'Dev', 'Esha', 'Farhan', 'Gauri', 'Hari'],
    'Age': [25.0, 42.0, np.nan, 31.0, 19.0, 56.0, np.nan, 28.0],
    'Department': ['HR', 'IT', 'Ops', 'IT', 'Marketing', 'Finance', 'Ops', 'HR'],
    'Salary': [50000.0, 85000.0, 62000.0, np.nan, 45000.0, 95000.0, np.nan, 52000.0],
    'Sample_Numbers': [12, 7, 22, 15, 40, 33, 8, 19],
    'Vowel_Test_Strings': ['apple', 'sky', 'education', 'rhythm', 'beautiful', 'xyz', 'ice', 'queue']
}
Employee_data=pd.DataFrame(data)
Employee_data
```
