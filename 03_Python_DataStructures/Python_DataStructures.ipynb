### 1.LIST- [ ]:
- List shows duplicates values
- 1 dimensional 1D
- stores Heterogenous datatype values
- Can be mutuable
- Indexing & Slicing can be done
  

```python
empty_cart=[]
type(empty_cart)
```

```python
Items_to_buy=["apple","grape","orange","pineapple,apple",6.75,0,9]
Items_to_buy
```

```python
type(Items_to_buy)
```

```python
print(Items_to_buy[-1]) #negative indexing
print(Items_to_buy[0]) #positive indexing

print(Items_to_buy[0:2]) #positive slicing
print(Items_to_buy[3:])

print(Items_to_buy[-6:-2]) #negative slicing 
print(Items_to_buy[::-1]) #reverse list using slicing

print(Items_to_buy[0::2])  #[start:stop:step]
print(Items_to_buy[0:len(Items_to_buy):1]) 
```

```python
Items_to_buy[1]="watermelon"
print(Items_to_buy[1]) #mutable
```

```python
Items_to_buy_2
```

### 3. DICTIONARIES {Key:Value}
- 2D
- Heterogenous
- Mutable
- Indexing and slicing is not possible
- Don't create duplicate key,but can create duplicate value for keys

```python
student = {
    "name": "Yuvasree",
    "age": 22,
    "score": 89
}
```

```python
student["city"] = "Erode"
print(student)

student.pop("score")
print(student)

```

```python
student = {
    "name": "Yuvasree",
    "age": 22
}

result = student.pop("score", None)

print(result)
```

```python
# dictionary.pop(key, default_value)
# key → the key you want to remove
# default_value → value returned if the key doesn't exist 

student = {
    "name": "Yuvasree",
    "age": 22,
    "score": 89
}

result = student.pop("score", None)

print(result)
print(student)
```

```python
# Key Doesn't Exist but no KeyError throwed since default value is given

student.pop("score",None)
print(student)
```

```python
student.keys()
```

```python
student.values()
```

```python
student.items()
```

```python
for key, value in student.items():
    print(key, value)
```

```python
cities = {"Chennai", "Erode", "Coimbatore"}
cities.add("Salem")
print(cities)
```

```python
nums = [1,2,2,3,4,4,5]

unique_nums = set(nums)
print(nums)
print(unique_nums)
```

```python
a = {1,2,3}

b = {3,4,5}

a.union(b)
```

```python
a.intersection(b)
```

### 4. TUPLES:

```python
location = (11.12, 78.65)
location[0]

location[1]
```

```python
lat, lon = location

print(lat)

print(lon)
```
