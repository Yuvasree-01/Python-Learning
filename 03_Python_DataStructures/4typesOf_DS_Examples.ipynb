## 1. LISTS (Highest Priority)

#### Create

```python
marks = [67, 87, 89, 90, 76]
```

#### Indexing

```python
marks[0]
marks[-1]
```

#### Slicing

```python
marks[1:4]
marks[:3]
marks[::2]
marks[::-1]
```

#### Add Elements

```python
marks.append(95)

marks.extend([96, 97])

marks.insert(2, 100)
```

#### Remove Elements

```python
marks.remove(87)

marks.pop()

marks.pop(1)
```

#### Useful Functions

```python
len(marks)

max(marks)

min(marks)

sum(marks)

sum(marks)/len(marks)  #gives average
```

#### Loop

```python
for mark in marks:
    print(mark)
```

## 2. DICTIONARIES (Highest Priority)

#### Create

```python
student = {
    "name": "Yuvasree",
    "age": 22,
    "score": 89
}
```

#### Access

```python
student["name"]

student["score"]
```

#### Add / Update

```python
student["city"] = "Erode"

student["score"] = 95
```

#### Delete

```python
del student["age"]

student.pop("score")
```

#### Get Keys

```python
student.keys()
```

#### Get Values

```python
student.values()
```

#### Get Both

```python
student.items()
```

#### Loop

```python
for key, value in student.items():
    print(key, value)
```

## 3. SETS

#### Create

```python
cities = {"Chennai", "Erode", "Coimbatore"}
```

#### Add

```python
cities.add("Salem")
```

#### Remove

```python
cities.remove("Erode")
```

#### Remove Duplicates

```python
nums = [1,2,2,3,4,4,5]

unique_nums = set(nums)

print(unique_nums)
```

#### Union

```python
a = {1,2,3}

b = {3,4,5}

a.union(b)
```

#### Intersection

```python
a.intersection(b)
```

## 4. TUPLES

#### Create

```python
location = (11.12, 78.65)
```

#### Access

```python
location[0]

location[1]
```

#### Unpacking

```python
lat, lon = location

print(lat)

print(lon)
```

#### Count

```python
t = (1,2,2,3)

t.count(2)
```

#### Index

```python
t.index(3)
```
