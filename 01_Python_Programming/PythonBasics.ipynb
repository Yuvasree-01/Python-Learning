# Number guessing game 

### level 1.Print if the user guessed number is right or not

```python
secret_number = 56
user_input = int(input("Enter you guess"))
if secret_number == user_input:
    print("Congratulations! you guessed the number")
else: 
    print("Ouch, Your guess is wrong")
```

### level 2.Print if the user guessed number is right or not but with max of 7 trails

##### Using for loop 

```python
secret_number = 56
for i in range(0,7):
    user_input = int(input("Enter you guess"))
    if secret_number == user_input:
        print("Congratulations! you guessed the number")
        break
    else: 
        print("Ouch, Your guess is wrong")
```

##### Using While loop

```python
secret_number = 56
while True:
    user_input = int(input("Enter you guess"))
    if secret_number == user_input:
        print("Congratulations! you guessed the number")
        break
    else: 
        print("Ouch, Your guess is wrong")
```

### Level 5. How to handle invalid values & characters

```python
import random

print("Welcome to Number Guessing Game!!")
print("Enter number betweeen 0 ---- 100")

secret_number = random.randint(0,100)
for i in range(0,7):
    try:
        if i!=6:
            user_input = int(input("Enter your guess:"))
            if  user_input<0 or user_input>100:
                print("Invalid number.Give number in range 0 to 100")
            if secret_number == user_input:
                print(f"Well Done! It took you {i+1} attempts to guess the number {secret_number}.")
                break
                    
            elif user_input < secret_number:
                print("The number is greater than" , user_input)
            elif user_input > secret_number:
                print("The number is less than" , user_input)
        else:
            print("Ouch, Your guess is wrong.You exceed your trails")
    except ValueError:
        print("Sorry, we take only numbers as input!")
            
```
